## Q A

### A.1)

```python
df_a = df[df['label'] != 2]
```

```python
x_a, y_a = df_a[:][['feature1', 'feature2']], df_a[:]['label']
```

```python
print(x_a.head())
print()
print(y_a.head())
print('-------------------')
print(x_a.info())
print()
print(y_a.info())
```

```python
X0 = df[df['label'] == 0][['feature1', 'feature2']]
X1 = df[df['label'] == 1][['feature1', 'feature2']]
```

```python
X0
```

```python
mean0 = np.mean(X0, axis=0)
mean1 = np.mean(X1, axis=0)

S0 = np.dot((X0 - mean0).T, (X0 - mean0))
S1 = np.dot((X1 - mean1).T, (X1 - mean1))
Sw = S0 + S1

w_fisher = np.dot(np.linalg.inv(Sw), (mean1 - mean0))
```

```python
def fisher(x):
    return (-w_fisher[0] * x) / w_fisher[1]
```

### A.2) 

```python
df_b = df[df['label'] != 2]
```

```python
X_b, Y_b = df_a[:][['feature1', 'feature2']], df_a[:]['label']
```

```python
x_b = df_b[:][['feature1', 'feature2']].reset_index(drop=True).to_numpy()
y_b = df_b[:]['label'].reset_index(drop=True).to_numpy()
```

```python
x_b[0]
```

```python
print(X_b.head())
print()
print(Y_b.head())
print('-------------------')
print(X_b.info())
print()
print(Y_b.info())
```

```python
y_b
```

```python
y_b = np.where(y_b == 0, -1, y_b)
```

```python
y_b
```

```python
w = np.zeros(x_b.shape[1])
b = 0
learning_rate = 0.01
epochs = 1000
```

```python
for epoch in range(epochs):
    for i in range(len(x_b)):
        if y_b[i] * (np.dot(w, x_b[i]) + b) <= 0:
            w += learning_rate * y_b[i] * x_b[i]
            b += learning_rate * y_b[i]
```

```python
def perceptron(x):
    return (-w[0] * x - b) / w[1]
```

### A.3)

```python
X_ls = np.c_[x_a, np.ones(x_a.shape[0])]

y_ls = np.where(y_a == 0, -1, y_a)

w_ls = np.dot(np.linalg.inv(np.dot(X_ls.T, X_ls)), np.dot(X_ls.T, y_ls))
```

```python
def least_squares(x):
    return (-w_ls[0] * x - w_ls[2]) / w_ls[1]
```

### Plot the result

```python
plt.scatter(X0[:]['feature1'], X0[:]['feature2'], color='red', label='Class 0')
plt.scatter(X1[:]['feature1'], X1[:]['feature2'], color='blue', label='Class 1')

x_vals_fisher= np.linspace(x_a[:]['feature1'].min(), x_a[:]['feature1'].max(), 100)
plt.plot(x_vals, fisher(x_vals_fisher), 'g--', label='Fisher')

x_vals_perceptron= np.linspace(X_b[:]['feature1'].min(), X_b[:]['feature1'].max(), 100)
plt.plot(x_vals, perceptron(x_vals_perceptron), 'b--', label='Perceptron')

x_vals_least_squares= np.linspace(x_a[:]['feature1'].min(), x_a[:]['feature1'].max(), 100)
plt.plot(x_vals, least_squares(x_vals_least_squares), 'r--', label='Least Squares')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

### Accuracy

```python
def accuracy(X, y, w, b=0):
    predictions = np.sign(np.dot(X, w) + b)
    correct = np.sum(predictions == y)
    return correct / len(y)
```

```python
fisher_acc = accuracy(x_a, y_a, w_fisher)
perceptron_acc = accuracy(x_a, y_b, w, b)
least_squares_acc = accuracy(x_a, y_ls, w_ls[:-1], w_ls[-1])
```

```python
accuracy_df = pd.DataFrame({
    "Classifier": ["Fisher", "Perceptron", "Least Squares"],
    "Accuracy": [fisher_acc, perceptron_acc, least_squares_acc]
})

print(accuracy_df)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv(FILE_PATH)
```

```python
df_a = df[df['label'] != 2]
x_a, y_a = df_a[:][['feature1', 'feature2']], df_a[:]['label']
```

```python
model = LogisticRegression()
model.fit(x_a, y_a)
```

```python
w = model.coef_[0]
b = model.intercept_[0]
print(w, b)
```

```python
def boundary(X):
    return (-w[0] * X - b) / w[1]
```

```python
print(y_a)
```

```python
print(x_a.head())
```

```python
plt.scatter(x_a[:50]['feature1'], x_a[:50]['feature2'], color='red', label='Class 0')
plt.scatter(x_a[50:]['feature1'], x_a[50:]['feature2'], color='blue', label='Class 1')

x_vals = np.linspace(x_a[:]['feature1'].min(), x_a[:]['feature1'].max(), 100)
plt.plot(x_vals, boundary(x_vals), 'c-', label='Logistic Boundary')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score

y_hat = model.predict(x_a)
acc = accuracy_score(y_a, y_hat)
print(f"Logistic Classifier Accuracy: {acc:.1f}")
```
