## Imports

```python
import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

## Read dataset

```python
FILE_PATH = "D:/M.A/T1/ML/Assignments/HW4_SVM  Kernel/sonar_csv.csv"
```

```python
df = pd.read_csv(FILE_PATH)
```

```python
df.head()
```

## Split data & label

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X
```

```python
y
```

## Covert labels into nums

```python
y = LabelEncoder().fit_transform(y)
```

```python
y
```

## Standardization

```python
X = StandardScaler().fit_transform(X)
```

```python
X
```

## Train/Test Split

```python
random = 13
results = []
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= random)
```

## Model

### Linear

#### Model

```python
linear_model = SVC(kernel= 'linear', random_state= random)
```

#### Trainning

```python
linear_model.fit(X_train, y_train)
```

#### Evaluation

```python
linear_train_acc = accuracy_score(y_train, linear_model.predict(X_train))
linear_test_acc = accuracy_score(y_test, linear_model.predict(X_test))
results.append(['Linear', linear_train_acc, linear_test_acc])
```

```python
results
```

### Polynomial

#### Model

```python
poly_model = SVC(kernel= 'poly', random_state= random)
```

#### Trainning

```python
poly_model.fit(X_train, y_train)
```

#### Eval

```python
poly_train_acc = accuracy_score(y_train, poly_model.predict(X_train))
poly_test_acc = accuracy_score(y_test, poly_model.predict(X_test))
results.append(['Polynomial', poly_train_acc, poly_test_acc])
```

```python
results
```

### RBF

#### Model

```python
rbf_model = SVC(kernel= 'rbf', random_state= random)
```

#### Train

```python
rbf_model.fit(X_train, y_train)
```

#### Eval

```python
rbf_train_acc = accuracy_score(y_train, rbf_model.predict(X_train))
rbf_test_acc = accuracy_score(y_test, rbf_model.predict(X_test))
results.append(['RBF', rbf_train_acc, rbf_test_acc])
```

```python
results
```

### Plot the result

```python
results_df = pd.DataFrame(results, columns=[
    'Kernel type', 'Train Set Classification Acc',
    'Test Set Classification Acc'
])
```
